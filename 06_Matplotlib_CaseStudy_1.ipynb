```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/Rajesh/Edureka/06_matplotlib/775_m6_datasets_v1.0/CityTemps.csv')
df['Year_Month'] = df['Year'].astype(str) + '_' + df['Month'].astype(str)
#display(df)
plt.figure(figsize=(8,4))
plt.hist(df['San Francisco'], color = 'black')
plt.xlabel('Temperature')
plt.title('San Francisco Weather')
plt.show()

plt.figure(figsize=(8,4))
plt.hist(df['Moscow'], color = 'skyblue')
plt.xlabel('Temperature')
plt.title('Moscow Weather')
plt.show()
```

```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
df = pd.read_csv('C:/Users/Rajesh/Edureka/06_matplotlib/775_m6_datasets_v1.0/Cars2015.csv')
#display(df)
df['Make'] = df['Make'].str.strip()
count_series = df.groupby(['Make'])['Model'].count()
df3 = count_series.to_frame(name=None).reset_index()
#display(df3)
labels = df3['Make'].tolist()
plt.figure(figsize=(16,16))
plt.pie(count_series, labels = labels)
plt.show()
```
