```python
import pandas as pd
from IPython.display import display
```

```python
df_ds = pd.read_csv("C:/Users/Rajesh/Edureka/05_numpy_pandas/DSScoreTerm1.csv")
df_math = pd.read_csv("C:/Users/Rajesh/Edureka/05_numpy_pandas/MathScoreTerm1.csv")
df_phys = pd.read_csv("C:/Users/Rajesh/Edureka/05_numpy_pandas/PhysicsScoreTerm1.csv")
display(df_ds)
display(df_math)
display(df_phys)

```

```python
df_ds = df_ds.drop(['Name','Ethinicity'], axis=1)
df_math = df_math.drop(['Name','Ethinicity'], axis=1)
df_phys = df_phys.drop(['Name','Ethinicity'],axis=1)
```

```python
df_ds['Score'] = df_ds['Score'].fillna(0)
df_math['Score'] = df_math['Score'].fillna(0)
df_phys['Score'] = df_phys['Score'].fillna(0)
```

```python
df_merge = pd.concat([df_ds,df_math,df_phys])
```

```python
display(df_merge)
```

```python
df_merge['Sex'] = df_merge['Sex'].str.replace('M','1')
df_merge['Sex'] = df_merge['Sex'].str.replace('F','2')
```

```python
display(df_merge)
df_merge.to_csv('ScoreFinalNew.csv', index=False)
```
